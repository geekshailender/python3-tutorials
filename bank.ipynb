```python
class Account():
  def __init__(self, owner,balance):
    self.owner = owner
    self.balance = balance
  def deposit(self, a):
     self.balance = self.balance + a
     return "Depoist Accepted"
  def withdraw(self,b):
    self.balance = self.balance - b
    if self.balance < 0:
      return "Insufficient Funds"
    else:
     return self.balance
```

```python
account = Account("Shailender",100)
```
